```python
# open a file called input.txt
with open('input1.txt') as f:
    # read the file into a list of lines
    lines = f.readlines()
    lines = [line.strip() for line in lines]
```

```python
elves = []
sum = 0

for line in lines:
    # print(line)
    if line != '':
        sum += int(line)
        # print(sum)
    else:
        elves.append(sum)
        sum = 0
# print the first 10 lines
elves.sort()
print(elves)   
```
